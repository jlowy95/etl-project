```python
# Dependencies
from bs4 import BeautifulSoup as bs
import requests
import time
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
from random import randint
```

```python
# URL of page to be scraped
search_term = 'skis'
url = f"https://offerup.com/search/?q={search_term}&radius=50"
print(url)
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=True, incognito=True,user_agent='Mozilla/5.0 (Windows NT 6.1; WOW64; rv:50.0) Gecko/20100101 Firefox/50.0')
```

```python
browser.visit(url)
button_xpath = '/html/body/div[1]/div/div[3]/div/div[2]/div[3]/button'
item_urls = []

for _ in range(40):
    try:
        browser.find_by_xpath(button_xpath).click()
    except:
        pass
    browser.execute_script("window.scrollTo(400, document.body.scrollHeight);")
    
html = browser.html
soup = bs(html, 'lxml')
   
all_item_urls = soup.find_all('a', {'class':['_109rpto db-item-tile', '_109rpto _1anrh0x']})
for item in all_item_urls:
    if '/item/' in item["href"]:
        item_url = f'https://offerup.com{item["href"]}'
        if item_url not in item_urls:
            item_urls.append(item_url)  




```

```python
print(len(item_urls))
for item in item_urls:
    print(item)
```

```python
listings_data = []

for item in item_urls:
    url = item
    browser.visit(url)
    html = browser.html
    soup = bs(html, 'lxml')
    item_title = getattr(soup.find('h1', class_='_t1q67t0 _1juw1gq'),'text',None)
    item_price = getattr(soup.find('span', class_='_ckr320'),'text',None)
    item_location = getattr(soup.find('a', class_='_g85abvs _133jvmu8'),'text',None)
    item_condition = getattr(soup.find('span', class_='_csmifkq'),'text',None)
    item_source = 'OfferUp'
    item_url = url
    listings_data.append({'item_title': item_title, 'item_price': item_price, 'item_location': item_location, 'item_condition': item_condition, 'item_url': item_url, 'item_source': item_source})
    time.sleep(randint(1,2))
```
